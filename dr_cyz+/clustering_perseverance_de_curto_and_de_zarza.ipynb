```python
import matplotlib.pyplot as plt
import cv2
import os
import random
import numpy as np
from PIL import Image
from sklearn.feature_extraction import image
from skimage import io
from sklearn.cluster import MiniBatchKMeans

!unzip '/content/drive/MyDrive/Colab Notebooks/cyz_1.1.zip'

data = []
folder = '/content/cyz_1.1/perseverance_cyz/png/'

for filename in os.listdir(folder):
    image = cv2.imread(os.path.join(folder,filename))
    if image is not None:
        image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)
        image = cv2.resize(image, (64,64))
        image = image.flatten()
        data.append([image, folder + filename])
```

```python
print(len(data))
features, images  = zip(*data)
```

```python
from sklearn.decomposition import PCA

features = np.array(features)
pca = PCA(2)
 
#Transform the data
df = pca.fit_transform(features)

n_cams = 22
kmeans = MiniBatchKMeans(n_clusters=n_cams)

#predict the labels of clusters.
label = kmeans.fit_predict(df)
 
#Getting unique labels
u_labels = np.unique(label)
```

```python
tx, ty = df[:,0], df[:,1]
tx = (tx-np.min(tx)) / (np.max(tx) - np.min(tx))
ty = (ty-np.min(ty)) / (np.max(ty) - np.min(ty))
```

```python
import matplotlib.pyplot
from matplotlib.pyplot import imshow

width = 4000
height = 3000
max_dim = 100

full_image = Image.new('RGBA', (width, height))
for img, x, y in zip(images, tx, ty):
    tile = Image.open(img)
    rs = max(1, tile.width/max_dim, tile.height/max_dim)
    tile = tile.resize((int(tile.width/rs), int(tile.height/rs)), Image.ANTIALIAS)
    full_image.paste(tile, (int((width-max_dim)*x), int((height-max_dim)*y)), mask=tile.convert('RGBA'))

matplotlib.pyplot.figure(figsize = (16,12))
imshow(full_image)
```

```python
import os
import json

full_image.save("/content/drive/MyDrive/Colab Notebooks/test_clustering_perseverance_cyz+.png")

path_clustering = "/content/drive/MyDrive/Colab Notebooks/test_clustering_perseverance_cyz+.json"

data = [{"path":os.path.abspath(img), "point":[float(x), float(y)]} for img, x, y in zip(images, tx, ty)]
with open(path_clustering, 'w') as outfile:
    json.dump(data, outfile)

print("saved K-means Clustering to %s" % path_clustering)
```
